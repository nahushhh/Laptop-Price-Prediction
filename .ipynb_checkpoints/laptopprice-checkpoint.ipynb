```python
def processor(text):
    if text == 'Intel Core i7' or text == 'Intel Core i5' or text == 'Intel Core i3':
        return text
    else:
        if text.split()[0] == 'Intel':
            return 'Other Intel Processor'
        else:
            return 'AMD Processor'
```

```python
df['CpuBrand'] = df['CpuName'].apply(processor)
```

```python
auto = df[[
    'Company', 'TypeName', 'Inches', 'Ram', 'OpSys', 'Weight', 'Price',
       'Touchscreen', 'Ips', 'CpuBrand', 'GpuBrand', 'SSD', 'HDD',
       'FlashStorage'
]]
```

Convert categorial columns to numeric columns.

```python
auto = pd.get_dummies(auto,
                      columns=[
                          'Company', 'TypeName','OpSys', 'CpuBrand', 'GpuBrand'
                      ],
                      drop_first=True)
```

```python
auto.head()
```

## Model Building

### Train Test Split

```python
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(auto,
                                     train_size=0.7,
                                     test_size=0.3,
                                     random_state=100)
```

Using Standard Scalar, we normalize the numeric columns.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
num_var = [
    'Inches', 'Ram', 'Weight', 'Price'
]

df_train[num_var] = scaler.fit_transform(df_train[num_var])
```

```python
df_train.head()
```

```python
y_train = df_train.pop('Price')
X_train = df_train
```

```python
X_train.head()
```

```python
X_train.shape
```

### RFE (Recursive feature elimination)
 A feature selection technique in machine learning that recursively eliminates less important features from a dataset based on their contribution to the model's performance, ultimately selecting the most relevant features for the model in a recursive manner.

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
rfe = RFE(estimator=lm, n_features_to_select=15)
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns, rfe.support_, rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

## Building model 

```python
X_train_rfe = X_train[col]
```

```python
# Adding a constant variable
import statsmodels.api as sm

X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lm = sm.OLS(y_train, X_train_rfe).fit() 
```

```python
print(lm.summary())
```

The p> |t| should be less than 0.05.
Therefore, we need drop columns like Company_Mediacom,Company_LG , Company_Google step wise.

## VIF

VIF stands for Variance Inflation Factor, which is a statistical measure used in regression analysis to assess the multicollinearity (correlation between predictor variables) in a set of data, with values above a certain threshold indicating high multicollinearity that may negatively impact the model's interpretability and reliability.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_rfe
vif['Features'] = X.columns
vif['VIF'] = [
    variance_inflation_factor(X.values, i) for i in range(X.shape[1])
]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by="VIF", ascending=False)
vif
```

The VIF score should be less than 5. 

```python
X_train_new1 = X_train_rfe.drop(["Company_Mediacom"], axis=1)

# Adding a constant variable
import statsmodels.api as sm

X_train_lm = sm.add_constant(X_train_new1)

lm = sm.OLS(y_train, X_train_lm).fit()  # Running the linear model

#Let's see the summary of our linear model
print(lm.summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_new1
vif['Features'] = X.columns
vif['VIF'] = [
    variance_inflation_factor(X.values, i) for i in range(X.shape[1])
]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by="VIF", ascending=False)
vif
```

```python
X_train_new2 = X_train_new1.drop(["Company_Google"], axis=1)

# Adding a constant variable
import statsmodels.api as sm

X_train_lm = sm.add_constant(X_train_new2)

lm = sm.OLS(y_train, X_train_lm).fit()  # Running the linear model

#Let's see the summary of our linear model
print(lm.summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_new2
vif['Features'] = X.columns
vif['VIF'] = [
    variance_inflation_factor(X.values, i) for i in range(X.shape[1])
]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by="VIF", ascending=False)
vif
```

## RFECV
RFECV stands for Recursive Feature Elimination with Cross-Validation, which is a feature selection technique in machine learning that combines recursive feature elimination with cross-validation to select the most important features while considering the model's performance on different subsets of data, improving the robustness of feature selection process.

```python
from sklearn.feature_selection import RFECV
from sklearn.linear_model import LinearRegression

# Init, fit
rfecv = RFECV(
    estimator=LinearRegression(),
    min_features_to_select=5,
    step=5,
    n_jobs=-1,
    scoring="r2",
    cv=5,
)

_ = rfecv.fit(X_train, y_train)
```

```python
X_train.columns[rfecv.support_]
```

```python
len(X_train.columns[rfecv.support_])
```

The model chose all the features.

```python
len(X_train.columns)
```

```python
X_train.loc[:, rfecv.support_]
```

```python
X_train_rfecv = X_train.loc[:, rfecv.support_]
```

```python
X_test_rfecv = X_test.loc[:, rfecv.support_]
```

## Linear Regression 
We use Linear Regression Model for calculating the R-squared score, Mean Absolute Error, Mean Squared Error and Root Mean Squared Error.

```python
lr = LinearRegression()
_ = lr.fit(X_train_rfecv, y_train)

print("Training R-sqaured:", lr.score(X_train_rfecv, y_train))
print("Testing R-squared:", lr.score(X_test_rfecv, y_test))
```

```python
lr = LinearRegression()
_ = lr.fit(X_train, y_train)

print("Training R-sqaured:", lr.score(X_train, y_train))
print("Testing R-squared:", lr.score(X_test, y_test))
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
print(mean_absolute_error(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
print(mean_squared_error(y_test, y_pred)**(1/2))
```
